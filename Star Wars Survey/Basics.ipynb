```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
print(star_wars.head(10))
```

```python
print(star_wars['RespondentID'].size)
star_wars.columns
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
print(star_wars['RespondentID'].size)
```

```python
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
print('\n')
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
true_false_one = {
    "Star Wars: Episode I  The Phantom Menace" : True,
    np.NaN : False
}
star_wars['Which of the following Star Wars films have you seen? Please select all that apply.'] = star_wars['Which of the following Star Wars films have you seen? Please select all that apply.'].map(true_false_one)
```

```python
true_false_two = {
    "Star Wars: Episode II  Attack of the Clones" : True,
    np.NaN : False
}
star_wars['Unnamed: 4'] = star_wars['Unnamed: 4'].map(true_false_two)
```

```python
true_false_three = {
    "Star Wars: Episode III  Revenge of the Sith" : True,
    np.NaN : False
}
star_wars['Unnamed: 5'] = star_wars['Unnamed: 5'].map(true_false_three)
```

```python
true_false_four = {
    "Star Wars: Episode IV  A New Hope" : True,
    np.NaN : False
}
star_wars['Unnamed: 6'] = star_wars['Unnamed: 6'].map(true_false_four)
```

```python
true_false_five = {
    "Star Wars: Episode V The Empire Strikes Back" : True,
    np.NaN : False
}
star_wars['Unnamed: 7'] = star_wars['Unnamed: 7'].map(true_false_five)
```

```python
true_false_six = {
    "Star Wars: Episode VI Return of the Jedi" : True,
    np.NaN : False
}
star_wars['Unnamed: 8'] = star_wars['Unnamed: 8'].map(true_false_six)
```

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
})
star_wars.columns
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6",
})
star_wars.columns
```

```python
star_wars.iloc[:,9:15].mean()
```

```python
star_wars.iloc[:,9:15].mean().plot.bar()

plt.title('Star Wars Movies Ranked')
plt.show()
```

## Star Wars Movies ranked

The highest ranked movie is the fifth movie, the lowest ranked is the third movie. The original trilogy is higher ranked than the prequels.

```python
star_wars.iloc[:,3:9].sum().plot.bar()
```

## Star Wars Movies seen

The most watched movie is the fifth movie, with the sixth movie coming next. The prequels were watched by less people who did the survey.

The most watched movie tends to have to best rating. The exception to this is the fourth movie having a better rating than the first movie despite being watched less.

## Separating survey population

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
sw_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == 'Yes']
not_sw_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == 'No']
```

```python
st_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == 'Yes']
not_st_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Trek franchise?"] == 'No']
```

```python
males.iloc[:,9:15].mean().plot.bar()
plt.title('Star Wars Movies Ranked For Males')
plt.show()
males.iloc[:,3:9].sum().plot.bar()
plt.title('Star Wars Movies Seen For Males')
plt.show()
```

```python
females.iloc[:,9:15].mean().plot.bar()
plt.title('Star Wars Movies Ranked For Females')
plt.show()
females.iloc[:,3:9].sum().plot.bar()
plt.title('Star Wars Movies Seen For Females')
plt.show()
```

When the data is separated by gender, there is a difference in the average ranking between the fourth and sixth movies.

```python
sw_fan.iloc[:,9:15].mean().plot.bar()
plt.title('Star Wars Movies Ranked For Star Wars Fans')
plt.show()
sw_fan.iloc[:,3:9].sum().plot.bar()
plt.title('Star Wars Movies Seen For Star Wars Fans')
plt.show()
```

```python
not_sw_fan.iloc[:,9:15].mean().plot.bar()
plt.title('Star Wars Movies Ranked For Non-Star Wars Fans')
plt.show()
not_sw_fan.iloc[:,3:9].sum().plot.bar()
plt.title('Star Wars Movies Seen For Non-Star Wars Fans')
plt.show()
```

Star Wars fans heavily prefer the Original Trilogy. Most of them have also see most of the movies.

Non-Star Wars fans on the other hand have mostly seen the fifth, sixth, and first movie. They tend to prefer the fifth and first movie.

The sample size for Non-Star Wars fans is lower which makes sense for a survey about Star Wars.
